## prepare dandelion dataset for diffusion model

```python
import numpy as np
import os
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
loc = r'/mnt/md0/tempFolder/nikhil/Diffusion/dandelion/'
saveLoc = r'/mnt/md0/tempFolder/nikhil/Diffusion/'
files = os.listdir(loc)
print(len(files), files[0:2])
```

```python
allImgs = []
for idx, file in enumerate(files):
    if 'jpg' not in file:
        continue
    img = mpimg.imread(loc + file)
    img = np.squeeze(np.array(img))
    imgShape=np.shape(img)
    if imgShape[0]==240 and imgShape[1]==240:
        #print(np.shape(img))
        allImgs.append(img)
    elif imgShape[0]>240 and imgShape[1]>240:
        img2 = img[imgShape[0]//2 - 120:imgShape[0]//2 + 120, imgShape[1]//2 - 120: imgShape[1]//2 + 120]
        #print(np.shape(img2))
        allImgs.append(img2)
    
```

```python
allImgs = np.asarray(allImgs)
np.save(saveLoc+'dandelion.npy', allImgs)
```
